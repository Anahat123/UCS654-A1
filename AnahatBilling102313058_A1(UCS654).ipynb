```python
import numpy as np
import pandas as pd
```

```python
# Roll number
r = 102313058

ar = 0.05 * (r % 7)
br = 0.3 * ((r % 5) + 1)

print("ar =", ar)
print("br =", br)
```

```python
df = pd.read_csv("/content/data-2.csv", encoding="latin1")
print(df.columns)
```

```python
#Step 1
x = df["no2"].dropna().values
z = x + ar * np.sin(br * x)
```

```python
#Step 2
mu = np.mean(z)
sigma_sq = np.mean((z - mu) ** 2)
lam = 1 / (2 * sigma_sq)
c = 1 / np.sqrt(2 * np.pi * sigma_sq)

print("Estimated Parameters:")
print("mu =", mu)
print("lambda =", lam)
print("c =", c)
```
